5. Why should data validation be performed during transformation rather than after loading?

- Performing data validation during the transformation phase (before loading) is crucial for several reasons:

Prevents Bad Data Propagation:
It stops invalid, incomplete, or inconsistent data from ever entering the target data warehouse or database. Once bad data is loaded, it can contaminate downstream systems, reports, and analyses, making it much harder and more costly to rectify.
Early Error Detection:
Identifying issues during transformation allows for problems to be caught closer to their source. This makes debugging easier and provides faster feedback to data source owners for correction.
Resource Efficiency:
Processing and storing invalid data in the target system wastes computational resources (storage, processing power, network bandwidth) and time. Validating early means only high-quality data consumes these resources.
Ensures Data Integrity:
It helps maintain the integrity and reliability of the target system. By enforcing rules and standards before data is committed, you ensure that the data aligns with business requirements and data models.
Minimizes Rollbacks and Rework:
If validation occurs only after loading, errors might necessitate rolling back entire loads, cleaning the data, and then re-loading. This is a time-consuming and disruptive process. Early validation reduces the likelihood of such scenarios.
Improved Performance: Loading only valid data can improve the performance of the loading process itself, as there are fewer anomalies to handle post-load.
Better Business Decisions:
High-quality data directly supports more accurate reporting, analytics, and ultimately, better business decisions.

7. Write an SQL query on Sales_Transactions to list all duplicate keys and their counts using the
business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount )

- SELECT
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount,
    COUNT(*) AS DuplicateCount
FROM
    Sales_Transactions
GROUP BY
    Customer_ID,
    Product_ID,
    Txn_Date,
    Txn_Amount
HAVING
    COUNT(*) > 1;

8. Enforcing Referential Integrity
Assume the following Customers_Master table:
Identify Sales_Transactions.Customer_ID values that violate referential integrity when joined with
Customers_Master and write a query to detect such violations

- Customer_ID values in Sales_Transactions that do not exist in Customers_Master are C105 and C106.Here the foreign key in a table (Sales_Transactions) does not match a valid primary key in the parent table (Customers_Master).

SELECT ST.Customer_ID
FROM Sales_Transactions ST
LEFT JOIN Customers_Master CM
ON ST.Customer_ID = CM.CustomerID
